# 3. A_total_normal 처리 후 TCP 통신 3개 미만 통신은 빼는 코드

```python
import pandas as pd
```

```python
oneway = pd.read_csv("test2_2nd_filter.csv", warn_bad_lines=False, error_bad_lines=False)
```

```python
oneway['SET']=oneway['ip.src'].str.cat(oneway['ip.dst'].astype(str),sep=" ")
```

```python
oneway
```

```python
small = oneway['SET'].value_counts()
```

```python
ssmall = small[small<3]
```

```python
ssmall #### 결과가 0이면 우리가 뽑은 attack과 매우 유사함을 의미 
```

```python
#### small의 값이 0이 나온경우 그만!!!!!!! 아래 코드 jmp
del oneway['SET']
oneway.to_csv("test2_1st_drop.csv",index=False, header=True)
```

```python
ssmall.index[0].split(' ')[0]
```

```python
a=[]
b=[]
for i in range(len(ssmall)):
    a.append(ssmall.index[i].split(' ')[0])
    b.append(ssmall.index[i].split(' ')[1])
```

```python
df = pd.DataFrame({"ip.src":a,"ip.dst":b})
```

```python
df
```

```python
merge1 = oneway.merge(df, on=['ip.src','ip.dst'], how="outer",suffixes=['','_'],indicator=True)
```

```python
merge1[merge1["_merge"]=="both"]
```

```python
merge1[merge1["_merge"]=="right_only"] # 없는거 확인
```

```python
handshake3 = merge1[merge1["_merge"]=="left_only"]
```

```python
handshake3
```

```python
handshake3.to_csv("test2_1st_drop.csv",index=False, header=True)
```
